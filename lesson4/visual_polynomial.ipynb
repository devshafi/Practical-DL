```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


# Polynomial function
def predict(x, w0, w1, w2):
    return w0 + w1 * x + w2 * (x**2)


# Plot function
@interact(w0=1.1, w1=1.1, w2=1.1)
def plot_curve(w0, w1, w2):
    x = np.linspace(-5, 5, 100)
    y = predict(x, w0, w1, w2)

    # plt.clf()
    plt.plot(x, y)
    # plt.axhline(0)
    # plt.axvline(0)
    plt.title(f"w0={w0}, w1={w1}, w2={w2}")
    plt.ylim(-5, 5)
    plt.show()


# ---- Try different weights ----
# plot_curve(0, 1, 0)    # straight line
# plot_curve(0, 0, 1)    # pure curve

# plot_curve(1, 1, 1)  # shifted curve
# plot_curve(0, -1, 0.5) # flipped + curve
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Polynomial function (cubic)
def predict(x, w0, w1, w2, w3):
    return w0 + w1 * x + w2 * (x**2) + w3 * (x**3)

# Plot function
@interact(w0=1.1, w1=1.1, w2=1.1, w3=1.1)
def plot_curve(w0, w1, w2, w3):
    x = np.linspace(-5, 5, 100)
    y = predict(x, w0, w1, w2, w3)

    plt.plot(x, y)
    plt.title(f"w0={w0}, w1={w1}, w2={w2}, w3={w3}")
    plt.ylim(-100, 100)
    plt.show()
```
