```python
# Drop irrelevent columns
df = df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1)

#check the results
df.head(10)
```

```python
# Convert boolean columns to integers (True -> 1, False -> 0)
boolean_cols = ['Pclass_2', 'Pclass_3', 'Embarked_Q', 'Embarked_S']
for col in boolean_cols:
    df[col] = df[col].astype(int)

# Re-check the head to confirm the conversion
df.head(10)
```

Normalize the Age Value


```python
# Find the maximum age in the dataset
max_age = df["Age"].max()
print(max_age)

df["Age_Normalized"] = df["Age"] / max_age
df.head()
```

```python
# Check for zeros in the 'Fare' column
zero_count = (df['Fare'] == 0).sum()

if zero_count > 0:
    print(f"✅ Yes, the 'Fare' column contains {zero_count} zero(s).")
else:
    print("✅ No zeros found in the 'Fare' column.")
```

```python
#remove Fare 0
df_cleaned = df[df['Fare'] != 0].copy()
df = df_cleaned
```

Scale the fare

```python
import numpy as np
df['Log_Fare'] = np.log10(df['Fare'])
df.head(10)
```

```python
# Separate feature(X) and target (y)
X = df.drop(['Survived','Age','Fare'], axis=1,)
y = df['Survived']
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
print(X)
```

```python
# Convert to PyTorch Tensors
X_tensor = torch.tensor(X.to_numpy(dtype="float32"))
y_tensor = torch.tensor(y.to_numpy(dtype="float32")).view(-1, 1)
```

```python
X = X_tensor
y = y_tensor

N, D = X.shape
H1, H2 = 16, 8   # two sets of neurons (hidden layers)

# 1) Parameters
W1 = (torch.randn(D, H1) * 0.01).requires_grad_()
b1 = torch.zeros(H1, requires_grad=True)

W2 = (torch.randn(H1, H2) * 0.01).requires_grad_()
b2 = torch.zeros(H2, requires_grad=True)

W3 = (torch.randn(H2, 1) * 0.01).requires_grad_()
b3 = torch.zeros(1, requires_grad=True)
```

```python
optimizer = torch.optim.SGD([W1, b1, W2, b2, W3, b3], lr=0.05)
```

```python
import torch.nn.functional as F

for epoch in range(10000):
    optimizer.zero_grad()

    z1 = X @ W1 + b1          # first linear value
    a1 = F.relu(z1)           # ReLU

    z2 = a1 @ W2 + b2         # second linear value
    a2 = F.relu(z2)           # ReLU

    preds = torch.sigmoid(a2 @ W3 + b3)
    loss = F.l1_loss(preds, y)

    loss.backward()           # gradients
    optimizer.step()          # gradient descent step

    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```
