```python
# load the dataset
import pandas as pd
df = pd.read_csv("train.csv")
```

```python
df.head(10)
```

```python
df.columns
```

```python
# Drop irrelevent columns
df = df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1)

#check the results
df.head(10)
```

```python
# one-hot encode the "Pclass" column
df = pd.get_dummies(df, columns=['Pclass'], drop_first=True)

df.head(10)
```

```python
# One-hot encode the 'Emberked' column
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
df.head(10)
```

```python
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
```

```python
df.head(10)
```

```python
df.isna().sum()
```

Applying inputation

```python
# Calculate median age
median_age = df['Age'].median()
print(median_age)
```

```python
df.fillna({'Age':median_age},inplace=True)
df.isna().sum()
```

```python
df.head(10)
```

```python
# Convert boolean columns to integers (True -> 1, False -> 0)
boolean_cols = ['Pclass_2', 'Pclass_3', 'Embarked_Q', 'Embarked_S']
for col in boolean_cols:
    df[col] = df[col].astype(int)

# Re-check the head to confirm the conversion
df.head(10)
```

Normalize the Age Value


```python
# Find the maximum age in the dataset
max_age = df["Age"].max()
print(max_age)

df["Age_Normalized"] = df["Age"] / max_age
df.head()
```

Scale the fare

```python
import numpy as np
df['Lof_Fare'] = np.log10(df['Fare'])
df.head(10)
```
